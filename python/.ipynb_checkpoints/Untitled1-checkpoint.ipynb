```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
# nicer looking default plots
plt.style.use('bmh')
from scipy import optimize as opt
from scipy import special as sp
from scipy.linalg import det
from scipy.misc import factorial
from itertools import compress
```

```python
def y1(a,x):
    sum1 = 1.
    fac1 = 1.
    N = 10
    for i in range(1,N):
        k1 = 4.*i-3.
        fac1 = fac1*(a+k1/2.)
        sum1 += fac1*x**(2.*i)/factorial(2.*i)
        
    return np.exp(-0.25*x**2)*sum1
    
def y2(a,x):
    sum2 = x
    fac2 = 1.
    N = 10
    for i in range(1,N):
        k2 = 4.*i-1.
        fac2 = fac2*(a+k2/2.)
        sum2 += fac2*x**(2.*i+1)/factorial(2.*i+1)
    
    return np.exp(-0.25*x**2)*sum2    
    
def dy1(a,x):
    sum1 = 1.
    dsum1 = 0.
    fac1 = 1.
    N = 10
    for i in range(1,N):
        k1 = 4.*i-3.
        fac1 = fac1*(a+k1/2.)
        sum1 += fac1*x**(2.*i)/factorial(2.*i)
        dsum1 += fac1*x**(2.*i-1)/factorial(2.*i-1)

    return np.exp(-0.25*x**2)*(-0.5*x*sum1 + dsum1)

def dy2(a,x):
    sum2 = x
    dsum2 = 1.
    fac2 = 1.
    N = 10
    for i in range(1,N):
        k2 = 4.*i-1.
        fac2 = fac2*(a+k2/2.)
        sum2 += fac2*x**(2.*i+1)/factorial(2.*i+1)
        dsum2 += fac2*x**(2.*i)/factorial(2.*i)

    return np.exp(-0.25*x**2)*(-0.5*x*sum2 + dsum2)

def ddy1(a,x):
    sum1 = 1.
    dsum1 = 0.
    ddsum1 = 0.
    fac1 = 1.
    N = 10
    for i in range(1,N):
        k1 = 4.*i-3.
        fac1 = fac1*(a+k1/2.)
        sum1 += fac1*x**(2.*i)/factorial(2.*i)
        dsum1 += fac1*x**(2.*i-1)/factorial(2.*i-1)
        ddsum1 += fac1*x**(2.*i-2)/factorial(2.*i-2)

    return np.exp(-0.25*x**2)*(-0.5*sum1 - x*dsum1 + ddsum1 + 0.25*x**2*sum1)

def ddy2(a,x):
    sum2 = x
    dsum2 = 1.
    ddsum2 = 0.
    fac2 = 1.
    N = 10
    for i in range(1,N):
        k2 = 4.*i-1.
        fac2 = fac2*(a+k2/2.)
        sum2 += fac2*x**(2.*i+1)/factorial(2.*i+1)
        dsum2 += fac2*x**(2.*i)/factorial(2.*i)
        ddsum2 += fac2*x**(2.*i-1)/factorial(2.*i-1)

    return np.exp(-0.25*x**2)*(-0.5*sum2 - x*dsum2 + ddsum2 + 0.25*x**2*sum2)

def U(a,x):
    Y1 = np.sqrt(np.pi)/(2**(0.5*a+0.25)*sp.gamma(3./4.+a/2.)*np.cos(np.pi*(0.25+a/2.)))*y1(a,x)
    Y2 = np.sqrt(np.pi)/(2**(0.5*a-0.25)*sp.gamma(1./4.+a/2.)*np.sin(np.pi*(0.25+a/2.)))*y2(a,x)
    return np.cos(np.pi*(0.25+a/2.))*Y1-np.sin(np.pi*(0.25+a/2.))*Y2

def dU(a,x):
    dY1 = np.sqrt(np.pi)/(2**(0.5*a+0.25)*sp.gamma(3./4.+a/2.)*np.cos(np.pi*(0.25+a/2.)))*dy1(a,x)
    dY2 = np.sqrt(np.pi)/(2**(0.5*a-0.25)*sp.gamma(1./4.+a/2.)*np.sin(np.pi*(0.25+a/2.)))*dy2(a,x)
    return np.cos(np.pi*(0.25+a/2.))*dY1-np.sin(np.pi*(0.25+a/2.))*dY2
    
def ddU(a,x):
    ddY1 = np.sqrt(np.pi)/(2**(0.5*a+0.25)*sp.gamma(3./4.+a/2.)*np.cos(np.pi*(0.25+a/2.)))*ddy1(a,x)
    ddY2 = np.sqrt(np.pi)/(2**(0.5*a-0.25)*sp.gamma(1./4.+a/2.)*np.sin(np.pi*(0.25+a/2.)))*ddy2(a,x)
    
    return np.cos(np.pi*(0.25+a/2.))*ddY1-np.sin(np.pi*(0.25+a/2.))*ddY2

def V(a,x):
    if np.isinf(sp.gamma(1./2.+a)):
        return np.inf
    return sp.gamma(1./2.+a)/np.pi*(np.sin(np.pi*a)*U(a,x)+U(a,-x))

def dV(a,x):
    return sp.gamma(1./2.+a)/np.pi*(np.sin(np.pi*a)*dU(a,x)+dU(a,-x))

def ddV(a,x):
    return sp.gamma(1./2.+a)/np.pi*(np.sin(np.pi*a)*ddU(a,x)+ddU(a,-x))
```

```python
#Hoppe fun:

def hoppeFun(E_,k,nu,delta):
    E = E_[0]
    
    if abs(E)>1:
        return 1
    elif np.isnan(E):
        print('Error: E =',E)
        return 1
    
    Ef = nu/(2*delta)
    Ep = Ef + E
    Em = Ef - E
    
    gamma_p = 1/(E+1j*(1-E**2))
    gamma_m = 1/(E-1j*(1-E**2))
    
    eta = np.sqrt(1-E**2)/Ef
    q_e = np.sqrt(1-k**2-1j*eta)
    q_h = np.sqrt(1-k**2+1j*eta)
    
    ap = -delta*Ep
    am = delta*Em
    xip = -np.sqrt(2/nu)*k
    xim = np.sqrt(2/nu)*k
    
    """
    chi_p = 1/(2**(-0.25-0.5*ap)*sp.gamma(0.25-ap/2.))*sp.pbdv(ap,xip)[0]
    chi_m = 1/(2**(-0.25-0.5*am)*sp.gamma(0.25-am/2.))*sp.pbdv(am,xim)[0]
    dchi_p = np.sqrt(2/nu)/(2**(-0.25-0.5*ap)*sp.gamma(0.25-ap/2.))*sp.pbdv(ap,xip)[1]
    dchi_m = np.sqrt(2/nu)/(2**(-0.25-0.5*am)*sp.gamma(0.25-am/2.))*sp.pbdv(am,xim)[1]
    
    """
    
    chi_p = U(-delta*Ep,-np.sqrt(2/nu)*k)
    chi_m = U(-delta*Em,np.sqrt(2/nu)*k)
    dchi_p = np.sqrt(2/nu)*dU(-delta*Ep,-np.sqrt(2/nu)*k)
    dchi_m = np.sqrt(2/nu)*dU(-delta*Em,np.sqrt(2/nu)*k)
    
    
    matrix = np.array([[chi_p, 0, gamma_p, gamma_m],
                       [0, chi_m, 1, 1],
                       [dchi_p, 0, 1j*gamma_p*q_e, -1j*gamma_m*q_h],
                       [0, dchi_m, 1j*q_e, -1j*q_h]])
    
    D = det(matrix)
    
    return D.imag

def hoppeFun2(E_,eF,X,b):
    e = E_[0]
    ky = X*b/(2*eF)
    
    ep = eF + e
    em = eF-e
    g = 1/(e+1j*np.sqrt(1-e*e))
    gc = 1/(e-1j*np.sqrt(1-e*e))
    lp = np.sqrt(1-ky**2+1j*np.sqrt(1-e*e)/eF)
    lm = np.sqrt(1-ky**2-1j*np.sqrt(1-e*e)/eF)
    
    ap = -ep/b
    am = -em/b
    xip = np.sqrt(2)*(-X)/np.sqrt(2*eF/b)
    xim = np.sqrt(2)*X/np.sqrt(2*eF/b)
    dxi = np.sqrt(2)/np.sqrt(2*eF/b)
    Ufactorp = 1/(2**(-0.25-0.5*ap)*sp.gamma(0.25-ap/2.))
    Ufactorm = 1/(2**(-0.25-0.5*am)*sp.gamma(0.25-am/2.))
    
    
    
    chip = Ufactorp*sp.pbdv(-ap-0.5,xip)[0]
    chim = Ufactorm*sp.pbdv(-am-0.5,xim)[0]
    dchip = dxi*Ufactorp*sp.pbdv(-ap-0.5,xip)[1]
    dchim = dxi*Ufactorm*sp.pbdv(-am-0.5,xim)[1]
    
    """
    
    chip = Ufactorp*U(ap,xip)
    chim = Ufactorm*U(am,xim)
    dchip = dxi*Ufactorp*dU(ap,xip)
    dchim = dxi*Ufactorm*dU(am,xim)
    
    """
    matrix = np.array([[chip, 0, g, gc],
                      [0, chim, 1, 1],
                      [dchip, 0, 1j*g*lm, -1j*gc*lp],
                      [0, dchim, 1j*lm, -1j*lp]])
    D = det(matrix)
    return D.imag
```
